## Case studies

### Synthetic dataset 

```python
from adspy_shared_utilities import plot_class_regions_for_classifier
```

```python
X_C2,y_C2 = make_classification(n_samples=100,
                                n_features=2,
                                n_redundant = 0,
                                n_informative=2,
                                n_clusters_per_class=1,
                                flip_y=0.1,
                                class_sep = 0.5,
                                random_state = 0)
```

```python
plt.scatter(X_C2[:,0],X_C2[:,1],c=y_C2)
plt.title('Synthetic dataset : Binary classification task')
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_C2,y_C2,random_state = 0)
```

```python
NBBinaryClassification = GaussianNB()
```

```python
NBBinaryClassification.fit(X_train,y_train)
```

```python
plot_class_regions_for_classifier(NBBinaryClassification,X_train,y_train,X_test,y_test,'Naive Bayes on synthetic binary data')
```

### Synthetic dataset (Nonlinearly seperable classes)

```python
X_D2,y_D2 = make_blobs(n_samples=100,
                       n_features=2,
                       centers=8,
                       cluster_std = 1.3,
                       random_state=4)

y_D2 = y_D2 % 2
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_D2,
                                                 y_D2,
                                                 random_state=0)
```

```python
NBNonLin = GaussianNB()
```

```python
NBNonLin.fit(X_train,y_train)
```

```python
plot_class_regions_for_classifier(NBNonLin,
                                  X_train,
                                  y_train,
                                  X_test,
                                  y_test,
                                  'Naive Bayes on nonlinearly separable synthetic data')
```

### Breast Cancer Dataset

```python
X_Cancer,y_Cancer = load_breast_cancer(return_X_y=True)
```

* Skipping its description etc since we have seen the dataset many times in earlier NBs

```python
X_train,X_test,y_train,y_test = train_test_split(X_Cancer,y_Cancer,random_state=0)
```

```python
NBCancer = GaussianNB()
```

```python
NBCancer.fit(X_train,y_train)
```

```python
print("Classification accuracy on training dataset = {:.2f}".format(NBCancer.score(X_train,y_train)))
print("Classification accuracy on test dataset = {:.2f}".format(NBCancer.score(X_test,y_test)))
```
